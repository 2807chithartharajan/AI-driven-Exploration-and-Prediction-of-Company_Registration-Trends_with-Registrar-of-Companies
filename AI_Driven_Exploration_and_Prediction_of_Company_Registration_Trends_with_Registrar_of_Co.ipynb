```python
# Load the data
df = pd.read_csv('Plantation_Companies.csv')
df.head(20)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
# Convert the 'Date of Registration' column to datetime
df['Date of Registraion'] = pd.to_datetime(df['DATE OF\nREGISTRATION'])
```

```python
# Extract the year from the date
df['Year'] = df['Date of Registraion'].dt.year

# Count the number of companies registered each year
registration_trends = df.groupby('Year').size()

# Plot the trends
plt.figure(figsize=(12,6))
plt.plot(registration_trends.index, registration_trends.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Companies Registered')
plt.title('Trend of Company Registrations')
plt.grid(True)
plt.show()
```
